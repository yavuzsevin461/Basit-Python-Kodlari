EXAMPLE::: kullanıcıdan öğrenci sayısını alıp, kullanıcının girmişolduğu öğrenci sayısına göre öğrencinin yaşlarını alıp, yas_liste sine kaydedip girilen yaşları ve ortalamasını yazan python kodunu yazınız.

```python
x=int(input("öğrenci sayısını giriniz"))
liste=[]
toplam=0
s=0
while (s<x):
    a=int(input("yaş giriniz"))
    toplam=toplam+a
    liste.append(a)
    s=s+1
ort=toplam/x
print(liste,toplam,ort)
```
